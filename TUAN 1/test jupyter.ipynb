```python
def solve():
    S = input()

    # Tạo chuỗi P chỉ gồm '0' và '1'
    P = ''.join(ch for ch in S if ch in '01')
    print(P)

    n = len(P)
    left = 0
    cnt0 = cnt1 = 0
    max_len = 0

    for right in range(n):
        if P[right] == '0':
            cnt0 += 1
        else:
            cnt1 += 1

        # Nếu lệch quá 1 → thu nhỏ cửa sổ
        while abs(cnt0 - cnt1) > 1:
            if P[left] == '0':
                cnt0 -= 1
            else:
                cnt1 -= 1
            left += 1

        # Cập nhật kết quả
        max_len = max(max_len, right - left + 1)

    print(max_len)
```
